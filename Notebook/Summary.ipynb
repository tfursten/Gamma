# Isolation-by-distance using Gamma Distribution

In previous IBD simulations, the Rayleigh, exponential, half-normal and triangular distributions each had a single parameter which was adjusted such that they each produced the same second moment of dispersal distance, $2\sigma^2$.  Here we apply the gamma distribution which is a two-parameter continuous distribution with a shape parameter, $\alpha$, and a rate parameter, $\beta$. The second moment of the gamma distribution is $\mathbb{E}[X^2]=\frac{\alpha(\alpha+1)}{\beta^2}$. If the $\sigma$ parameter and an $\alpha$ parameter are provided at the start of the simulation then $\beta=+\sqrt{\frac{\alpha(1+\alpha)}{2\sigma^2}}$. If the $\sigma$ parameter and $\beta$ parameter are provided then $\alpha$ is any positive solution to the quadratic equation with $a=1, b=1, c=-2\sigma^2 \beta^2$. 

````python
import numpy as np
import math
import pandas as pd

sigma = 1.0

def alpha(b,s):
    t1 = math.sqrt(1-4*(-2*s*s*b*b))
    return (-1+t1)/2

def beta(a,s):
    return math.sqrt((a*(1+a))/(2*s*s))

a_values = np.array([1,2,4,8,10])
b_values = [beta(a,sigma) for a in a_values]
#b_values = np.array([0.25,0.5,1,2,4,10])
#a_values = [alpha(b,sigma) for b in b_values]
ex = [a_values[i]/b_values[i] for i in range(len(a_values))]
ex2 = [(a_values[i]*(a_values[i]+1))/b_values[i]**2 for i in range(len(a_values))]
var = [a_values[i]/b_values[i]**2 for i in range(len(a_values))]
sim = []
for i in range(len(a_values)):
    x = np.random.gamma(a_values[i],1/float(b_values[i]),1000000)
    sim.append(np.mean(x))
df = pd.DataFrame.from_items([('$\sigma$',[sigma for i in range(len(a_values))]),\
                              ('$$2\sigma^2$$',[2*sigma**2 for i in range(len(a_values))]),\
                              ('alpha',a_values),('beta',b_values),\
                              ('Expected$$E[X]$$',ex),('Variance$$Var[X]$$',var),\
                              ('Second Moment$$E[X^2]$$',ex2),('Simulated $$\hat{\mu}$$',sim)])
df
````

##Visualization of dispersal

````python
import random
from matplotlib import colors
from matplotlib import cm

def xy2i(x,y,mx,my):
    return x*my+y

def i2xy(i,mx,my):
    return (i/my,i%my)

def wrap_around(x,w):
    return x%w

def disperse(x,y,d,mx,my):
    a = random.random()*2.0*math.pi
    dx = math.floor(d*math.cos(a)+x+0.5)
    dy = math.floor(d*math.sin(a)+y+0.5)
    newX = wrap_around(dx,mx)
    newY = wrap_around(dy,my)
    return xy2i(newX,newY,mx,my)

for k,fn in enumerate(f_list):
    infile = open(fn,'r')
    #maxX = math.ceil(maxval[k]*2)
    #maxY = math.ceil(maxval[k]*2)
    maxX = 10
    maxY = 10
    X = 0.5*maxX
    Y = 0.5*maxY
    pop = np.zeros((maxX*maxY))
    infile.readline()
    for line in infile:
        d = float(line.strip().split('\t')[0])
        i = disperse(X,Y,d,maxX,maxY)
        pop[i] += 1
    pop = np.array([i for i in pop])
    pop = pop.reshape((maxX,maxY))
    fig,ax = plt.subplots(figsize=(5,5))
    cax = ax.imshow(pop, interpolation='nearest',cmap=cm.binary)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    plt.tick_params(axis='x',which='both',bottom='off',top='off',labelbottom='off',labelsize=6)
    plt.tick_params(axis='y',which='both',right='off',left='off',labelleft='off',labelsize=6)
    #cbar = fig.colorbar(cax)
    #plt.savefig("map_"+fn+".png", dpi=200)
    plt.show()

````

````python
from fractions import Fraction
files = !ls gamma_*[0-9].txt
files.insert(4, files.pop(1))
burn = 10
avg_ibd = []
avg_f = []
avg_s = []
outfile = open("gamma_summary.txt",'w')
outfile.write("dist\tsigma\tibd\tibdcorr\tr\n")
for fn in files:
    f = open(fn,'r')
    for line in range(burn+1):
        f.next()
    data = np.array(np.genfromtxt(f,usecols=range(7,51+6),dtype=object,unpack=True))
    f.seek(0)
    for line in range(burn+1):
        f.next()
    shat, fhat = np.array(np.genfromtxt(f,usecols=(1,5), unpack=True))
    avg_f.append(np.mean(fhat))
    avg_s.append(np.mean(shat))
    x,y = data.shape
    total_ibd = np.zeros(y)
    data2 = np.zeros((x,y))
    for i in xrange(x):
        for j in xrange(y):
            data[i][j] = Fraction(data[i][j])
            data2[i][j] = float(data[i][j])
    mean1 = np.mean(data2,axis=1)
    for i in range(y):
        row = data[:,i]
        num = 0
        den = 0
        for j in row:
            num += j.numerator
            den += j.denominator
        total_ibd[i] = num/float(den)
    for i in range(x):
        for j in range(y):
            if total_ibd[j] == 1:
                data2[i][j] = data2[i][j] - total_ibd[j]
            else:
                data2[i][j] = (data2[i][j]-total_ibd[j])/float(1-total_ibd[j])
    mean2 = np.mean(data2,axis=1)
    for i,v in enumerate(mean1):
        outfile.write(fn.strip('.txt').capitalize()+'\t'+"1"+'\t'+str(v)+'\t'+str(mean2[i])+'\t'+str(i+1)+"\n")
    avg_ibd.append(np.mean(total_ibd))
    f.close()
outfile.close()
f2 = open("dem_summary.txt",'w')
f2.write("Distribution,E[Sigma],Avg Sigma,Avg f,Avg IBD\n")
for i in range(len(avg_f)):
    f2.write(files[i]+",1,"+str(avg_s[i])+','+str(avg_f[i])+','+str(avg_ibd[i])+'\n')
f2.close()
````
